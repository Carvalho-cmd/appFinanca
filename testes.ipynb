```python
import pandas as pd
```

```python
from supabase import create_client, Client
from dotenv import load_dotenv
import os
```

```python
load_dotenv()

url = os.getenv("supabase_url")
key = os.getenv("supabase_key")

supabase: Client = create_client(url,key)
```

```python
try:
    despesas = supabase.table("tb_despesas").select("*").execute()
except Exception as e:
    print(e)
```

```python
despesas = pd.DataFrame(despesas.data)
```

```python
cartoes = [
    'Nubank',
    'Mercado Pago',
    'Santander',
    'XP',
]
```

```python
cadastro_despesa = {
    'Descriçao': ['Macbook'],
    'Balanço': ['05/2026'],
    'Categoria': ['Parcelado'],
    'Data despesa': ['2026-05-30'],
    'Cartão': ['Nubank'],
    'Local': [''],
    'Responsável': [''],
    'Valor': [5437,22],
    'Qtd. Parcelas': [24]
}
```

```python
despesa = {
    'descricao': [],
    'balanco': [],
    'categoria': [],
    'data_despesa': [],
    'cartao': [],
    'local': [],
    'responsavel': [],
    'valor': [],
    'parcela': []
}
```

```python
data = cadastro_despesa['Data despesa'][0]
mes = int(data[5:7])
ano = int(data[0:4])

valor = cadastro_despesa['Valor'][0]
int_qtd_parcelas = int(cadastro_despesa['Qtd. Parcelas'][0])
valor_parcelado = valor/int_qtd_parcelas

for i in range(cadastro_despesa['Qtd. Parcelas'][0]):

    parcela = i + 1

    if mes == 13:
        mes = 1 
        ano += 1

    despesa['descricao'].append(cadastro_despesa['Descriçao'][0])

    if mes < 10:
        despesa['balanco'].append(f'0{mes}/{ano}')
    else:
        despesa['balanco'].append(f'{mes}/{ano}')

    despesa['categoria'].append(cadastro_despesa['Categoria'][0])
    despesa['data_despesa'].append(cadastro_despesa['Data despesa'][0])
    despesa['cartao'].append(cadastro_despesa['Cartão'][0])
    despesa['local'].append(cadastro_despesa['Local'][0])
    despesa['responsavel'].append(cadastro_despesa['Responsável'][0])
    despesa['valor'].append(f'{valor_parcelado: .2f}')
    despesa['parcela'].append(f'{parcela}/{cadastro_despesa['Qtd. Parcelas'][0]}')

    mes += 1

```

```python
df_despesa = pd.DataFrame(despesa)

#display(df_despesa)
```

```python
df_ano = despesas[despesas["balanco"].str.contains(ano)]
```

```python
df_despesa_total_por_ano = df_ano[["balanco","valor"]].groupby("balanco").sum().reset_index()

datas = pd.to_datetime(df_despesa_total_por_ano["balanco"], format="%m/%Y")

df_despesa_total_por_ano["balanco"] = datas.dt.strftime("%B")

display(df_despesa_total_por_ano)
```

```python
mapa_meses = {
    "01": "Janeiro", "02": "Fevereiro", "03": "Março", "04": "Abril",
    "05": "Maio", "06": "Junho", "07": "Julho", "08": "Agosto",
    "09": "Setembro", "10": "Outubro", "11": "Novembro", "12": "Dezembro"
}

df_despesa_total_por_ano = df_ano[["balanco", "valor"]].groupby("balanco").sum().reset_index()

df_despesa_total_por_ano["balanco"] = df_despesa_total_por_ano["balanco"].str[:2].map(mapa_meses)

display(df_despesa_total_por_ano)
```
